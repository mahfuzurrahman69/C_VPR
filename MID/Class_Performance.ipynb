```python
import os
import random
import cv2
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAIN_DIR = '/content/drive/MyDrive/CIFAR-10-images-master/test'
CATEGORIES = []
for c in os.listdir(TRAIN_DIR):
    CATEGORIES.append(c)
print(CATEGORIES)
```

```python
TEST_DATA = []

# Iterate through the categories and images
for category in CATEGORIES:
    path = os.path.join(TEST_DIR, category)
    class_num = CATEGORIES.index(category)
    for img_file in os.listdir(path):
        img_arr = cv2.imread(os.path.join(path, img_file))
        if img_arr is not None:
            TEST_DATA.append([img_arr, class_num])
print(len(TEST_DATA))
```

```python
plt.figure(figsize=(20, 10))
for i in range(10):
    if i < len(TEST_DATA):
        plt.subplot(2, 5, i + 1)
        plt.imshow(cv2.cvtColor(TEST_DATA[i][0], cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(f'Class: {CATEGORIES[TEST_DATA[i][1]]}')
plt.show()
```
